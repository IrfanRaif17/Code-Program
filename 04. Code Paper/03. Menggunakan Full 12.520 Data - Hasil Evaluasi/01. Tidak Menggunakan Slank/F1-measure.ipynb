```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('gabung_all.csv')
```

```python
# Menghitung jumlah label dengan masing-masing sentimennya
sentimen_count = data['label'].value_counts()

# Menampilkan hasilnya
print(sentimen_count)

# Membuat pie chart untuk setiap label
fig, ax = plt.subplots()
ax.pie(sentimen_count, labels=sentimen_count.index, autopct='%1.1f%%')
ax.axis('equal')
plt.title('Jumlah Label Data Asli \n(total = 12.520 tweets)')
plt.savefig('hasil F1 Measure/pie_chart_label_asli.png')
plt.show()
```

```python
label_dict = {'neutral': 0, 'positive': 1, 'negative': -1}
data['label'] = data['label'].apply(lambda x: label_dict[x])

# Menampilkan DataFrame setelah pengubahan label
print(data)
```

```python
label_dict = {'neutral': 0, 'positive': 1, 'negative': -1}
data_clus['polarity'] = data_clus['polarity'].apply(lambda x: label_dict[x])

# Menampilkan DataFrame setelah pengubahan label
print(data_clus)
```

```python
y_true = data['label']
```

```python
y_pred = data_clus['polarity']
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score

# Menghitung akurasi
accuracy = accuracy_score(y_true, y_pred)

# Menghitung recall
# Menghitung recall dengan rata-rata 'macro'
recall = recall_score(y_true, y_pred, average='macro')

# Menghitung presisi dengan rata-rata 'macro'
precision = precision_score(y_true, y_pred, average='macro')

# Menghitung F1 score
f1 = f1_score(y_true, y_pred, average = 'macro')

accuracy_percent = round(accuracy * 100, 2)
print("Accuracy (%):", accuracy_percent)

recall_percent = format(recall * 100, ".2f")
print("Recall (%):", recall_percent)

precision_score = format(precision * 100, ".2f")
print("Precision score (%):", precision_score)

f1_percent = format(f1 * 100, ".2f")
print("F1 score (%):", f1_percent)
```
