# Evaluasi Matrix 3 * 3 Macro
Dibawah ini evaluasi matrix weighted tidak menggunakan library

```python
import pandas as pd
import numpy as np

# Fungsi untuk membaca file CSV dan menghitung matriks evaluasi
def calculate_metrics_from_csv(file_path):
    # Baca data dari file CSV
    data = pd.read_csv(file_path)
    
    # Asumsi kolom label asli bernama 'label' dan kolom label prediksi bernama 'polarity'
    actual = data['label']
    predicted = data['polarity']
    
    # Definisikan kelas
    classes = ['negative', 'neutral', 'positive']
    class_indices = {cls: idx for idx, cls in enumerate(classes)}
    
    # Inisialisasi confusion matrix
    confusion_matrix = np.zeros((3, 3), dtype=int)
    
    # Isi confusion matrix
    for act, pred in zip(actual, predicted):
        confusion_matrix[class_indices[act], class_indices[pred]] += 1
    
    # Hitung metrik evaluasi untuk setiap kelas
    metrics = {}
    precision_list = []
    recall_list = []
    f1_list = []
    
    for idx, cls in enumerate(classes):
        TP = confusion_matrix[idx, idx]
        FP = confusion_matrix[:, idx].sum() - TP
        FN = confusion_matrix[idx, :].sum() - TP
        TN = confusion_matrix.sum() - (TP + FP + FN)
        
        # Precision
        precision = TP / (TP + FP) * 100 if (TP + FP) > 0 else 0
        precision_list.append(precision)
        
        # Recall
        recall = TP / (TP + FN) * 100 if (TP + FN) > 0 else 0
        recall_list.append(recall)
        
        # F1 Score
        f1 = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
        f1_list.append(f1)
        
        metrics[cls] = {
            'Recall': recall,
            'Precision': precision,
            'F1-score': f1
        }
    
    # Accuracy keseluruhan
    accuracy = np.trace(confusion_matrix) / confusion_matrix.sum() * 100
    
    # Rata-rata makro untuk semua kelas
    macro_avg = {
        'Accuracy': accuracy,
        'Recall': np.mean(recall_list),
        'Precision': np.mean(precision_list),
        'F1-score': np.mean(f1_list),
    }
    
    return confusion_matrix, macro_avg, metrics

# Path ke file CSV
file_path = './hasil2stem/Hasil_tweets_negative.csv'

# Hitung metrik evaluasi dari file CSV
confusion_matrix, macro_avg, metrics = calculate_metrics_from_csv(file_path)

# Cetak confusion matrix
print("Confusion Matrix:")
print(confusion_matrix)

# Cetak metrik evaluasi untuk setiap kelas
for cls in metrics:
    print(f"\nMetrics for {cls.capitalize()}:")
    print(f"Recall: {metrics[cls]['Recall']:.2f}%")
    print(f"Precision: {metrics[cls]['Precision']:.2f}%")
    print(f"F1-score: {metrics[cls]['F1-score']:.2f}%")

# Cetak rata-rata makro
print(f"\nMacro Average:")
print(f"Accuracy: {macro_avg['Accuracy']:.2f}%")
print(f"Recall: {macro_avg['Recall']:.2f}%")
print(f"Precision: {macro_avg['Precision']:.2f}%")
print(f"F1-score: {macro_avg['F1-score']:.2f}%")
```

# Evaluasi Matrix 3 * 3 Weighted
Dibawah ini evaluasi matrix weighted tidak menggunakan library

```python
import pandas as pd
import numpy as np

# Fungsi untuk membaca file CSV dan menghitung matriks evaluasi
def calculate_metrics_from_csv(file_path):
    # Baca data dari file CSV
    data = pd.read_csv(file_path)
    
    # Asumsi kolom label asli bernama 'label' dan kolom label prediksi bernama 'polarity'
    actual = data['label']
    predicted = data['polarity']
    
    # Definisikan kelas
    classes = ['negative', 'neutral', 'positive']
    class_indices = {cls: idx for idx, cls in enumerate(classes)}
    
    # Inisialisasi confusion matrix
    confusion_matrix = np.zeros((3, 3), dtype=int)
    
    # Isi confusion matrix
    for act, pred in zip(actual, predicted):
        confusion_matrix[class_indices[act], class_indices[pred]] += 1
    
    # Hitung metrik evaluasi untuk setiap kelas
    metrics = {}
    for idx, cls in enumerate(classes):
        TP = confusion_matrix[idx, idx]
        FP = confusion_matrix[:, idx].sum() - TP
        FN = confusion_matrix[idx, :].sum() - TP
        TN = confusion_matrix.sum() - (TP + FP + FN)
        
        # Precision
        precision = TP / (TP + FP) * 100 if (TP + FP) > 0 else 0
        
        # Recall
        recall = TP / (TP + FN) * 100 if (TP + FN) > 0 else 0
        
        # F1 Score
        f1 = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
        
        metrics[cls] = {
            'Recall': recall,
            'Precision': precision,
            'F1-score': f1
        }
    
    # Accuracy keseluruhan
    accuracy = np.trace(confusion_matrix) / confusion_matrix.sum() * 100
    
    # Rata-rata tertimbang untuk semua kelas
    weighted_avg = {
        'Accuracy': accuracy,
        'Recall': sum([metrics[cls]['Recall'] * confusion_matrix[class_indices[cls], :].sum() for cls in classes]) / confusion_matrix.sum(),
        'Precision': sum([metrics[cls]['Precision'] * confusion_matrix[class_indices[cls], :].sum() for cls in classes]) / confusion_matrix.sum(),
        'F1-score': sum([metrics[cls]['F1-score'] * confusion_matrix[class_indices[cls], :].sum() for cls in classes]) / confusion_matrix.sum(),
    }
    
    return confusion_matrix, weighted_avg, metrics

# Path ke file CSV
file_path = './hasil2stem/Hasil_tweets_negative.csv'

# Hitung metrik evaluasi dari file CSV
confusion_matrix, weighted_avg, metrics = calculate_metrics_from_csv(file_path)

# Cetak confusion matrix
print("Confusion Matrix:")
print(confusion_matrix)

# Cetak metrik evaluasi untuk setiap kelas
for cls in metrics:
    print(f"\nMetrics for {cls.capitalize()}:")
    print(f"Recall: {metrics[cls]['Recall']:.2f}%")
    print(f"Precision: {metrics[cls]['Precision']:.2f}%")
    print(f"F1-score: {metrics[cls]['F1-score']:.2f}%")

# Cetak rata-rata tertimbang
print(f"\nWeighted Average:")
print(f"Accuracy: {weighted_avg['Accuracy']:.2f}%")
print(f"Recall: {weighted_avg['Recall']:.2f}%")
print(f"Precision: {weighted_avg['Precision']:.2f}%")
print(f"F1-score: {weighted_avg['F1-score']:.2f}%")
```
