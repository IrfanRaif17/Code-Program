```python
def accuracy_score_manual(y_true, y_pred):
    correct = sum(yt == yp for yt, yp in zip(y_true, y_pred))
    return correct / len(y_true)

def precision_recall_f1_manual(y_true, y_pred, cls):
    tp = sum(yt == yp == cls for yt, yp in zip(y_true, y_pred))
    fp = sum(yp == cls and yt != cls for yt, yp in zip(y_true, y_pred))
    fn = sum(yt == cls and yp != cls for yt, yp in zip(y_true, y_pred))

    precision = tp / (tp + fp) if tp + fp != 0 else 0
    recall = tp / (tp + fn) if tp + fn != 0 else 0
    f1 = 2 * precision * recall / (precision + recall) if precision + recall != 0 else 0

    return precision, recall, f1

def macro_metric(y_true, y_pred, metric_func):
    unique_classes = set(y_true)
    metrics_sum = 0
    for cls in unique_classes:
        metric_value = metric_func(y_true, y_pred, cls)
        metrics_sum += metric_value

    return metrics_sum / len(unique_classes)

def weighted_metric(y_true, y_pred, metric_func):
    unique_classes = set(y_true)
    class_counts = {cls: y_true.count(cls) for cls in unique_classes}
    total = len(y_true)

    weighted_sum = 0
    for cls in unique_classes:
        metric_value = metric_func(y_true, y_pred, cls)
        weighted_sum += (class_counts[cls] / total) * metric_value

    return weighted_sum
```

```python
# Ensuring y_true and y_pred are lists
if not isinstance(y_true, list):
    y_true = y_true.tolist()
if not isinstance(y_pred, list):
    y_pred = y_pred.tolist()

# Menghitung akurasi
accuracy = accuracy_score_manual(y_true, y_pred)

# Menghitung presisi, recall, dan F1-score weighted
weighted_precision = weighted_metric(y_true, y_pred, lambda yt, yp, cls: precision_recall_f1_manual(yt, yp, cls)[0])
weighted_recall = weighted_metric(y_true, y_pred, lambda yt, yp, cls: precision_recall_f1_manual(yt, yp, cls)[1])
weighted_f1 = weighted_metric(y_true, y_pred, lambda yt, yp, cls: precision_recall_f1_manual(yt, yp, cls)[2])

# Konversi ke persen
accuracy_percent = round(accuracy * 100, 2)
print("Accuracy (%):", accuracy_percent)

weighted_recall_percent = format(weighted_recall * 100, ".2f")
print("Weighted Recall (%):", weighted_recall_percent)

weighted_precision_percent = format(weighted_precision * 100, ".2f")
print("Weighted Precision score (%):", weighted_precision_percent)

weighted_f1_percent = format(weighted_f1 * 100, ".2f")
print("Weighted F1 score (%):", weighted_f1_percent)
```
